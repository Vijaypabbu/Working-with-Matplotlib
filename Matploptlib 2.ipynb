```python
#data from external sources as csv files
import numpy as np
import csv 
from collections import Counter #to count no of times a word seen in dataset 
plt.style.use('fivethirtyeight')
with open('Desktop\data\data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    #row =next(csv_reader): to select 1st row values in the form of dict in csv file
    #print(row):to print that row
    #print(row['LanguagesWorkedWith'].split(';')):to print values in lww seperated by ;
    #to counts every word that seen in the each row of lww:
    language_counter = Counter()
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))
#print(language_counter.most_common(15)):1st 15 items are printed 
    language = []
    popularity = []
    for item in language_counter.most_common(15):
        language.append(item[0])
        popularity.append(item[1])
#print(language)
#print(popularity)
language.reverse()#to make desc order of list to asc
popularity.reverse()#to make desc order of list to asc
plt.barh(language,popularity,label='languages with popularity')#barh for bar chart in horizonatal view
plt.xlabel('languages')
plt.ylabel('popularity')
plt.title('Most Popular Languages')
#plt.tight_layout()
plt.legend()


```

```python
#plotting live data in realtime
import random
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation   #automatically append the value for every interval

plt.style.use('fivethirtyeight')

x_vals = []
y_vals = []
index = count()#used to count each time when the value came
def animate(i):# used to append the value in between(0,5) for every interval set by ani
    x_vals.append(next(index))
    y_vals.append(random.randint(0, 5))
    plt.plot(x_vals,y_vals)
anim =FuncAnimation(plt.gcf(),animate,interval=1000)#get current figure,fuction,interval for callin the function 

plt.tight_layout()
plt.show()
```
